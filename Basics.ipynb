```python
# Cleaning the column names
new_col = []

def change_case(str):
    s1 = re.sub('(.)([A-Z][a-z]+)', r'\1_\2', str)
    return re.sub('([a-z0-9])([A-Z])', r'\1_\2', s1).lower()

def col_cleaner(col):
    col = col.replace("yearOfRegistration", "registration_year")
    col = col.replace("monthOfRegistration", "registration_month")
    col = col.replace("notRepairedDamage", "unrepaired_damage")
    col = col.replace("dateCreated", "ad_created")
    col = change_case(col)
    return col

```

```python
for i in autos.columns:
    new_col.append(col_cleaner(i))
    
autos.columns = new_col
```

```python
autos['odometer'] = autos['odometer'].str.replace('km', '')
autos['odometer'] = autos['odometer'].str.replace(',', '')
autos['price'] = autos['price'].str.replace(',', '')
autos['price'] = autos['price'].str.replace('$', '')
autos['odometer'] = autos['odometer'].astype(int)
autos['price'] = autos['price'].astype(int)
autos.rename({'odometer': 'odometer_km'}, axis =1, inplace = True)

autos.describe(include = 'all')
```

We will now be aggregating throught the top 20 most common car brands from the dataset and calculating their mean listed price.

```python
brands = autos['brand'].value_counts().index[:20]
mean_price = {}
```

```python
for b in brands:
    rows = autos[autos['brand'] == b]
    mean = rows['price'].mean().round()
    mean_price[b] = mean
```

```python
mean_price = pd.Series(mean_price)
mean_price.sort_values()
```

As we can see from the data, although volkswagen is by far the most popular car brand, its mean price is far from the cheapest. Fiats are the cheapest listed car on average with a price of 2837. Sonstige autos are the most expensive, with an average price of 12785. 

```python
mean_mileage = {}

for b in brands:
    rows = autos[autos['brand'] == b]
    mean = rows['odometer_km'].mean().round()
    mean_mileage[b] = mean
```

```python
mean_mileage = pd.Series(mean_mileage)

df = pd.DataFrame(mean_price, columns = ['mean_price'] )
df['mean_mileage'] = mean_mileage
```
